```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
city_data = "Resources/cities.csv"
df = pd.read_csv(city_data)
df.head()
```

```python
max_temp = []
for mt in df["Max Temp"]:
    max_temp.append(mt)
    
lat = []
for lt in df["Lat"]:
    lat.append(lt)

humidity = []
for hmd in df["Humidity"]:
    humidity.append(hmd)
    
cloudy = []
for cld in df["Cloudiness"]:
    cloudy.append(cld)
    
wind_speed = []
for ws in df["Wind Speed"]:
    wind_speed.append(ws)
    
```

```python
plt.scatter(lat,max_temp)
plt.grid()
plt.title("City Latitude vs. Max Temp (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
```

```python
plt.scatter(lat,humidity)
plt.grid()
plt.title("City Latitude vs. Humidity (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
```

```python
plt.scatter(lat,cloudy)
plt.grid()
plt.title("City Latitude vs. Cloudiness (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
```

```python
plt.scatter(lat,wind_speed)
plt.grid()
plt.title("City Latitude vs. Wind Speed (01/05/17)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
```
